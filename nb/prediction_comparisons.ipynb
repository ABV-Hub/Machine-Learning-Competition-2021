```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pred1 = pd.read_csv('../data/data-driven_pancakes_submission_1.csv')
pred2 = pd.read_csv('../data/datadrivenpancakes_submission_2.csv')
pred2v2 = pd.read_csv('../data/datadrivenpancakes_submission_2v2.csv')
pred3 = pd.read_csv('../data/datadrivenpancakes_submission_3.csv')
```

```python
fig = plt.figure(figsize=(20,8))

ax1 = fig.add_subplot(2,1,1)
ax1.plot(pred1.index, pred1['DTC'], 'b', label='pred1')
ax1.plot(pred2.index, pred2['DTC'], 'g', label='pred2')
ax1.plot(pred3.index, pred3['DTC'], 'r', label='pred3')
ax1.set_xlabel('index')
ax1.set_ylabel('DTC')
ax1.invert_yaxis()
ax1.legend()

ax2 = fig.add_subplot(2,1,2)
ax2.plot(pred1.index, pred1['DTS'], 'b', label='pred1')
ax2.plot(pred2.index, pred2['DTS'], 'g', label='pred2')
ax2.plot(pred3.index, pred3['DTS'], 'r', label='pred3')
ax2.set_xlabel('index')
ax2.set_ylabel('DTS')
ax2.invert_yaxis()
ax2.legend()

plt.show()
```
